```python
import pandas as pd

url = "https://drive.google.com/file/d/1g3uhw_y3tboRm2eYDPfUzXXsw8IOYDCy/view?usp=sharing"
path = 'https://drive.google.com/uc?export=download&id='+url.split('/')[-2]

to_drop = ['PassengerId', 'Name', 'Sex', 'Ticket', 'Cabin', 'Embarked']
data = pd.read_csv(path).drop(columns=to_drop)
```

#### Exercise 0: First we need to split the target(y) from the features(X)

```python
X = data.copy()
y = data.pop('Survived')
```

#### Exercise 1: Split the data into training and test sets using 80% for training and 20% for testing. Set `random_state=42`

```python
from sklearn.model_selection import train_test_split
from sklearn.dummy import DummyClassifier


train_X, test_X, train_y, test_y = train_test_split(X, y, random_state=42, test_size=0.2)
baseline = DummyClassifier(strategy='most_frequent')
baseline.fit(train_X, train_y)
baseline.score(test_X, test_y)
```
